2. Weather Data Analyzer
a. Analyze weather data (temperature, humidity, etc.) for a given region
b. Use variables and data types to store weather data
c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)
d. Use functions to calculate average temperature and humidity, and print
results


```python
# Define variables to store weather data
temperature = [25, 28, 30, 22, 20, 18]
humidity = [60, 70, 80, 50, 40, 30]
weather_conditions = []

# Function to calculate average temperature
def average_temperature(temperature):
    return sum(temperature) / len(temperature)

# Function to calculate average humidity
def average_humidity(humidity):
    return sum(humidity) / len(humidity)

# Function to determine weather conditions
def determine_weather(temperature, humidity):
    if temperature > 25 and humidity < 60:
        return "Sunny"
    elif temperature < 20 and humidity > 70:
        return "Rainy"
    else:
        return "Cloudy"

# Determine weather conditions for each day
for i in range(len(temperature)):
    weather_conditions.append(determine_weather(temperature[i], humidity[i]))

# Print results
print("Weather Data Analysis")
print("---------------------")
print(f"Average Temperature: {average_temperature(temperature)}Â°C")
print(f"Average Humidity: {average_humidity(humidity)}%")
print("Weather Conditions:")
for i in range(len(weather_conditions)):
    print(f"Day {i+1}: {weather_conditions[i]}")
```
